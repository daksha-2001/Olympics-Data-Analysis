```python
def fetch_medal_tally(year,country):
    flag=0
    if year=='Overall' and country=='Overall':
        df=medal_tally
     
    if year=='Overall' and country!='Overall':
        flag=1
        df=medal_tally[medal_tally['region']==country]
    
    if year!='Overall' and country=='Overall':
        df=medal_tally[medal_tally['Year']==int(year)]
    
    if year!='Overall' and country!='Overall':
        df=medal_tally[(medal_tally['Year']==int(year)) & (medal_tally['region']==country)]
     
    if flag==1:
        df= df.groupby("Year").sum()[['Gold','Silver','Bronze']].sort_values('Year').reset_index()
    else:
        df= df.groupby("region").sum()[['Gold','Silver','Bronze']].sort_values('region').reset_index()
    df['Total']=df['Gold']+df['Silver']+df['Bronze']
    return df

fetch_medal_tally('Overall',"India")
```

```python
def most_successful(df,sport):
    temp_df=df.dropna(subset=['Medal'])
    
    if sport!='Overall':
        temp_df=temp_df[temp_df['Sport']==sport]
        
    x=temp_df['Name'].value_counts().reset_index().merge(df,left_on="index",right_on="Name",how='left')[["index","Name_x","Sport","region"]].drop_duplicates(subset=['index'])
    x.rename(columns={'index':'Name',"Name_x":"Medals"},inplace=True)
    return x.head(20)

most_successful(df,'Swimming')
```
